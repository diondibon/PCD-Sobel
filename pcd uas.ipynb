```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
def sobelOperator(img):
    gradient = np.copy(img)
    size = gradient.shape
    for x in range(1, size[0] - 1):
        for y in range(1, size[1] - 1):
            gx = (img[x - 1][y + 1] + 2*img[x][y + 1] + img[x + 1][y + 1]) - (img[x - 1][y - 1] + 2*img[x][y - 1] + img[x + 1][y - 1])
            gy = (img[x - 1][y - 1] + 2*img[x - 1][y] + img[x - 1][y + 1]) - (img[x + 1][y - 1] + 2*img[x + 1][y] + img[x + 1][y + 1])
            gradient[x][y] = min(255, np.sqrt(gx**2 + gy**2))
    return gradient
    pass
```

```python
img = cv2.imread('jenis koi.jpeg')
cv2.imshow('original',img)
```

```python
imgOri = cv2.imread("ikan.jpg")
```

```python
#gambar awal
plt.title('Citra awal')
plt.imshow(imgOri)
plt.show()
```

```python
#Konversi ke grayscale
img = cv2.cvtColor(imgOri, cv2.COLOR_BGR2GRAY)
plt.title('Hasil grayscale')
plt.imshow(img)
plt.show()
```

```python
#Konversi ke metode sobel
img = sobelOperator(img)
plt.title('Hasil metode sobel')
plt.imshow(img)
plt.show()
```

```python
#Hasil Konversi sobel ke grayscale
plt.title('Hasil konversi sobel ke grayscale')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```
